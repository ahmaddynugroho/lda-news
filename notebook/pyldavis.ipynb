```python
args = {'a': 0.005624704431183458, 'e': 0.6836874433905906, 'k': 10.0}
model = LdaMulticore(
    corpus=corpus,
    id2word=dictionary,
    num_topics=args['k'],
    alpha=args['a'],
    eta=args['e'],
    per_word_topics=True,
    workers=3,
    random_state=99
)
```

```python
with open('../dist/model.pickle', 'wb') as f:
    pickle.dump(model, f)
```
