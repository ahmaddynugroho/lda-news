```python
import pandas as pd

df = pd.read_csv('../dist/tuning.csv')
```

```python
x = []
y = []

for i in range(11, 21):
    x.append(i)
    y.append(df[df['k'] == i]['c'].max()) 
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(x, y)
plt.xlabel('Jumlah Topik (k)')
plt.ylabel('Coherence Score')
```

```python
fig.savefig('../dist/tuning_plot.svg')
```
